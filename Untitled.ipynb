```python
# 기본 / EDA
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# 데이터 불러오기
import os
import tarfile
import urllib.request

# 데이터 분리
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
```

# 데이터 불러오기

```python
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/rickiepark/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()

def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, "housing.csv")
    return pd.read_csv(csv_path)
```

# 데이터 탐색

```python
# 데이터프레임 탐색
fetch_housing_data()
housing = load_housing_data()
housing.head()
```

```python
# 데이터 타입
housing.info()

# 데이터 타입 중 object 탐색
housing['ocean_proximity'].value_counts()
```

```python
# 데이터 정량화 파악
housing.describe()
```

```python
# 데이터 시각화
housing.hist(bins=50,figsize=(10,10))
plt.show()
```

```python
housing['median_income'].hist()
```

```python
# 범주화 적용
housing['income_cat'] = pd.cut(housing['median_income'],
                              bins=[0., 1.5, 3.0, 4.5, 6., np.inf], # float으로 전부 설정
                              labels=[1,2,3,4,5]) 
```

```python
# income_cat 확인
print(housing['income_cat'].value_counts())
housing['income_cat'].hist() # 정규분포에 비슷하게 나눠짐.
```

# 학습 / 검증용 데이터 분리

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
# n_splits - 검증 레이어 개수 / test_size = 검증용 데이터 분리 비율
```

```python
# StratifiedShuffleSplit 중 split Method는 train/test의 인덱스들 을 반환합니다.
# for문으로 train / test 의 index를 반환합니다.
# split의 파라미터값으로 ( X : 학습데이터, y : 타겟 데이터)
# X에 타겟을 포함하는 것이 분리가 잘됨 (그후  np.zeros(n_samples)로 타겟값을 대체합니다.)

for train_index, test_index in split.split(housing, housing['income_cat']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

```python
# 분리 검증 - 타겟 열의 비율로 검증
print('< S_test_set >\n',strat_test_set['income_cat'].value_counts() / len(strat_test_set))
print('\n')
print('< housing >\n', housing['income_cat'].value_counts() / len(housing))

# 비율이 동일하게 나온것으로 확인. 분리가 제대로 잘 됐다.
```

# 데이터 탐색 & 시각화

```python
housing = strat_train_set.copy()
```
