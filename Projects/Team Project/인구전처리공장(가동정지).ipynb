```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')

전처리파일 = '서울특별시1'
```

```python
dd = pd.read_csv(f'New1/{전처리파일}.csv',index_col=0,encoding='cp949')
dd =dd.reset_index(drop=False)
dd['읍면동별(1)']. unique()
dd = dd[['읍면동별(1)','읍면동별(2)','2018.1']]
dd = dd.rename(columns={'읍면동별(1)':'시군구명','읍면동별(2)':'읍/면/동','2018.1':'등록인구수'})
읍면동 = [index for index, row in dd[dd['읍/면/동']=='소계'].iterrows()]
불필요시군구 = [index for index, row in dd[dd['읍/면/동']=='읍면동별(2)'].iterrows()]
dd = dd.drop(읍면동)
dd = dd.drop(불필요시군구)
dd['시도명'] = f'{전처리파일}'
dd = dd[['시도명','시군구명','읍/면/동','등록인구수']]
dd.to_csv(f'Dataset/human/{전처리파일}.csv',index=0)
data = pd.read_csv(f'Dataset/human/{전처리파일}.csv')
data
```
