```python
import pandas as pd
pd.set_option('display.max_row', 500)
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import re
def dong(x):
    import re

    p = re.compile('[가-힣]\S*[읍면동가]')

    r = p.findall(x)
    
    return ''.join(r)

import re
def dong1(x):
    import re

    p = re.compile('[가-힣]\S*[면동가]')

    r = p.findall(x)
    
    return ''.join(r)
검색할시도명='강원도'
검색할시군구명='횡성군'

제공기관명 = f'{검색할시도명} {검색할시군구명}'

빼올ui = 검색할시도명
검색할구 = 검색할시군구명
PP = pd.read_csv('Dataset/전국인구밀도포함.csv')
um = PP[PP['시군구명']==검색할구]
u1 = um['읍/면/동'].tolist()
u1
len(u1)
```

```python
df= pd.read_csv('dataset/전국어린이보호구역표준데이터.csv',encoding='cp949')

# df = df.rename(columns=df.iloc[0])
# df= df.drop(df.index[0])

df1 = df.loc[df.loc[:,'제공기관명'].str.contains(제공기관명,na=False)]

detail_1 = df1[df1['제공기관명']==제공기관명]


detail_1['읍/면/동'] = detail_1['소재지지번주소'].apply(lambda x : dong(x))


a= 제공기관명.split(" ")[0]
b= 제공기관명.split(" ")[1]

detail_1['시도명']= a
detail_1['시군구명']= b

a = detail_1[detail_1['읍/면/동']=='']
# a['읍/면/동'] = a['읍/면/동'].replace("",'거진읍')
detail_1[detail_1['읍/면/동']==''] = a
child = detail_1[['시도명','시군구명','읍/면/동']]
child['어린이보호구역개수'] = 1
child[child['읍/면/동']=='']
child
detail_1.head(1)
```

```python
group = child.groupby(['시도명','시군구명','읍/면/동'])
a = group.sum()
a = a.reset_index()
child = a
```

```python
print(len(child))
print(len(um))
display(child)
display(um)
```

```python
df = pd.merge(um,child,on=['시도명','시군구명','읍/면/동'],how='left')
```
