```python
data_null = data[(data['소재지지번주소'].isnull()==1) & (data['시군구명']=='서구')]
```

```python
import requests
import json
import numpy as np

addrs = list(data_null['소재지도로명주소'])
locations = []


for addr in addrs:
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(addr)
    headers = {
        "Authorization" : "KakaoAK ac29d0b5ff83ba31f85e0d6f9f38e797"}
    place = requests.get(url, headers = headers).json()['documents']
    locations.append(place)
    
city = []

for i in range(len(locations)):
    
    try:
        city.append(locations[i][0].get('address').get('address_name'))
       
        
    except IndexError:
        print(i,'번째 주소 못가져옴',end ='')
        print()
        city.append('없음')
#         town.append('없음')
        
    except AttributeError:
        city.append(locations[i][0].get('road_address').get('region_2depth_name'))
        


city_town = list(city)
# df_temp = pd.DataFrame(city_town, columns = ['2', '3'])
# asan['소재지지번주소'] = df_temp[['3']]
# asan['소재지지번주소'] = df_temp[['3']]
```
