```python
# import modules
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-dark')
%config InlineBackend.figure_format
pd.options.display.max_rows = 20
pd.options.display.max_columns = 20
```

```python
# load data 
raw_data = pd.read_csv('../Data/FIFA.csv')
#copy the raw_data
fd = raw_data.copy()
fd.head(2)
```

```python
# df 정보 확인
fd.info()
```

```python
#결측값 확인
fd.isnull().sum()
# 결측값 없음
```

    변환처리 : Height ft/in -> cm

```python
def t_cm(x):
    a,b = map(int,x.split("'")) # ' 문자열 그대로 받아오기 \' (" ' ") -> 크게 상관x
    return a*30.48 + b*2.54
fd['Height_cm'] = fd.Height.apply(lambda x: t_cm(x))
```

```python
fd.drop('Height',axis=1,inplace=True)
```

```python
fd.head(2)
```

```python
#이상치값 확인
sns.boxplot(x=fd['Height_cm'])
fd['Height_cm'].unique()
```

    변환처리 : Weight (lb->kg)

```python
fd["Weight_kg"] = fd["Weight_lb"] * 0.4535924
fd["Weight_kg"] = fd["Weight_kg"].round(2)
fd["Weight_kg"]
fd.drop('Weight_lb',axis=1,inplace=True)
```

```python
fd.head(2)
```

```python
#이상치 값 확인
sns.boxplot(x=fd['Weight_kg'])
fd["Weight_kg"].unique()
```

```python
# 의미 없는 columns 제거
# ID / Jersey_Number / Nationality
```

```python
fd.drop(['ID','Jersey_Number','Nationality'],axis=1,inplace=True)
```

```python
fd.head()
```

```python
#Preferred_Foot
fd["Preferred_Foot"].unique()
fd["Preferred_Foot"] = fd["Preferred_Foot"].replace({'Left':0,'Right':1}) 
fd.head(2)
```

```python
fd.corr()
```

```python
#Work_Rate
fd["Work_Rate"].unique()
# Work_Rate = AWR (Attack_work_rate) + DWR (Defensive_work_rate)
fd[['AWR','DWR']] = fd["Work_Rate"].str.split('/',1,expand=True)
fd['DWR']= fd['DWR'].str.lstrip()
fd[['AWR','DWR']].head(2)
fd['AWR']=fd['AWR'].replace({'Low':0,'Medium':1,'High':3})
fd['DWR']=fd['DWR'].replace({'Low':0,'Medium':1,'High':3})
fd[['AWR','DWR']].head(1)
```

```python
fd.drop('Work_Rate',axis=1,inplace=True)
```

```python
fd.head(2)
```

```python
#age 범주화\
fd['Age'].min() #16
fd['Age'].max() #45

fd.groupby('Age')['Age'].count()

fd['Age'].unique()
```

```python
sns.histplot(x=fd['Age'])
```

```python
wage_av_c = fd[['Club','Wage']].groupby('Club').mean().round(2)
wage_av_c['avW'] = wage_av_c["Wage"]
wage_av_c.drop('Wage',axis=1,inplace=True)
```

```python
wage_av_c.sort_values('avW',ascending=False)
wage_av_c.reset_index(inplace=True)
```
