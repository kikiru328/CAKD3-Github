*subset : 기준이되는 컬럼 명을 넣음

*drop duplicates(subset)

```python
import pandas as pd
import numpy as np
arr = np.array([1,2,1,2,2,3,3,3])
data = arr.reshape(4,2)
df = pd.DataFrame(data, columns = ['a','b'])
df
```

```python
df1 = df.copy()
df1.drop_duplicates()
```

```python
df2 = df.copy()
df2.drop_duplicates(subset=['b']) #default = first
```

```python
df3 = df.copy()
df3.drop_duplicates(subset='b',keep='last')
```

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['sex'] = titanic['sex'].replace({'male':0,'female':1})
```

```python
titanic.rename(columns={'sex':'gender','class':'c_level'},inplace=True)
```

```python
titanic.rename(columns=str.lower,inplace=True)
```

```python
df = titanic[['survived','pclass','gender','c_level']]
new_order = ['pclass','gender','c_level','survived']
df = df[new_order]
```

```python
# df['gender'].replace({0:'male',1:'female'},inplace=True)
df['gender'].replace([1,0],['female','male'],inplace=True)
```

```python
#encoding?
#male/female -> encoding. 숫자로 바꿔줌
# Label enconding
# 희소행렬? 01 -> one hot encoding
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
features = ['gender','c_level']
for feature in features:
    df[feature] = le.fit_transform(df[feature]) #인코딩 실행됨. -학습: fit함수를 사용
df.head()
```

```python
df.gender.value_counts()
```

```python
df.c_level.value_counts()
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.keys())
boston.target
boston.DESCR
boston.feature_names
df_boston = pd.DataFrame(boston.data,columns=boston.feature_names)
df_boston['target'] = boston.target
df_boston.head()
```

```python
# https://huidea.tistory.com/39
# 스케일링 정규화
# 데이터의 분포가 가우시안 분포가 아닐 경우
# 모든 값들을 0 ~ 1 사이로 바꾸어줌
# 데이터 분포를 비교 확인할 수 있음 
# 단위가 달라도 공평하게 0 ~ 1 사이로 있으니까.
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_boston['RM'] = scaler.fit_transform(df_boston[['RM']])
df_boston.head()
```

```python
# 표준화
# 평균이 0 이고 분산이 1 인  표준 정규 분포로 변환한다는 것  (가우시안 표준 분포)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_boston['INDUS'] = scaler.fit_transform(df_boston[['INDUS']])
df_boston.head() #평균을 0으로 만들어줌
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.figure(figsize=(10,8))
plt.style.use('seaborn-dark')
```

```python
sns.distplot(df_boston[['target']])
```

```python
# 로그 변환
# 정규분포가 이상적임. ---정규분포 가정 (회기분석?) = 모수통계로 한대 근데 그게뭐지
# 
import numpy as np
df_boston['target'] = np.log1p(df_boston[['target']])
```

```python
#환원
df_boston['target'] = np.expm1(df_boston[['target']])
df_boston.head()
```

```python
students = [('jack', 34, 'Sydeny'),
            ('Riti', 30, 'Delhi'),
            ('Aadi', 16, 'New York'),
            ('Riti', 30, 'Delhi'),
            ('Riti', 30, 'Delhi'),
            ('Riti', 30, 'Mumbai'),
            ('Aadi', 40, 'London'),
            ('Sachin', 30, 'Delhi')
            ]
```

```python
dfObj = pd.DataFrame(students, columns=['Name', 'Age', 'City'])
```

```python
dfObj
```

```python
duplicateRowsDF = dfObj[dfObj.duplicated()] # except first
duplicateRowsDF
```

```python
duplicateRowsDF = dfObj[dfObj.duplicated(keep='last')] # except last
duplicateRowsDF 
```

```python
duplicateRowsDF = dfObj[dfObj.duplicated(['Age', 'City'])]
duplicateRowsDF
```
