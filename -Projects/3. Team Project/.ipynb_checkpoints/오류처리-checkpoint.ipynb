```python
data_null = data[(data['소재지지번주소'].isnull()==1) & (data['시군구명']=='남구')]
```

```python
import requests
import json
import numpy as np

addrs = list(data_null['소재지도로명주소'])
locations = []
need_to_find = []

for addr in addrs:
    url = 'https://dapi.kakao.com/v2/local/search/address.json?query={}'.format(addr)
    headers = {
        "Authorization" : "KakaoAK ac29d0b5ff83ba31f85e0d6f9f38e797"}
    place = requests.get(url, headers = headers).json()['documents']
    locations.append(place)
    
city = []

for i in range(len(locations)):
    
    try:
        city.append(locations[i][0].get('address').get('address_name'))
       
        
    except IndexError:
        print(i,'번째 주소 못가져옴',end ='')
        print()
        city.append('없음')
        
        
        need_to_find.append(i)
#         town.append('없음')
        
    except AttributeError:
        city.append(locat=2ions[i][0].get('road_address').get('region_2depth_name'))
        


city_town = list(city)
# df_temp = pd.DataFrame(city_town, columns = ['2', '3'])
# asan['소재지지번주소'] = df_temp[['3']]
# asan['소재지지번주소'] = df_temp[['3']]
```

```python
또찾아야됨 = []
아예없음 = []
for 번호 in need_to_find:
    try:
        import selenium
        from selenium import webdriver
        from selenium.webdriver.common.keys import Keys
        options = webdriver.ChromeOptions()
        options.headless = True
        options.add_argument('window-size-1920x1080')
        url = 'https://www.naver.com/'
        driver = webdriver.Chrome('C:/Users/kikir/anaconda3/chromedriver.exe',options=options)
        # driver = webdriver.Chrome('D:/Tools/chromedriver.exe')
        driver.get(url)
        a = data_null.iloc[int(f'{번호}')]['소재지도로명주소']
        search_box = driver.find_element_by_xpath('//*[@id="query"]')
        search_box.send_keys(a)
        search_box.send_keys(Keys.RETURN)
        get =driver.find_element_by_xpath('//*[@id="no-matched-address-list"]').text
        # uv_st
        city[int(f'{번호}')] = get
        print(get)
    
    except NoSuchElementException:
        또찾아야됨.append(번호)
        print(번호, data_null.iloc[int(f'{번호}')]['소재지도로명주소'],'\n')
        
for 재번호 in 또찾아야됨:
    try:
        import selenium
        from selenium import webdriver
        from selenium.webdriver.common.keys import Keys
        options = webdriver.ChromeOptions()
        options.headless = True
        options.add_argument('window-size-1920x1080')
        url = 'https://www.naver.com/'
        driver = webdriver.Chrome('C:/Users/kikir/anaconda3/chromedriver.exe',options=options)
        # driver = webdriver.Chrome('D:/Tools/chromedriver.exe')
        driver.get(url)
        a = data_null.iloc[int(f'{재번호}')]['소재지도로명주소']
        search_box = driver.find_element_by_xpath('//*[@id="query"]')
        search_box.send_keys(a)
        search_box.send_keys(Keys.RETURN)
        if driver.find_element_by_xpath('//*[@id="unique"]').text != None:
            get =driver.find_element_by_xpath('//*[@id="unique"]').text
            city[int(f'{재번호}')] = get
            print(get)
            
    except NoSuchElementException:
        아예없음.append(재번호)
        print(재번호, data_null.iloc[int(f'{번호}')]['소재지도로명주소'])
        
        
    
```
