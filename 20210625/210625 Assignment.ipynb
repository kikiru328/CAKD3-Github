```python
import random
answer = list('가나다라마바사아자차카타파하')
with open("Info_name.txt",'w') as info_name:
    for i in range(1000):
        name = random.choice(answer) + random.choice(answer)
        height = random.randrange(140,201)
        weight = random.randrange(40,151)
        bmi = round(int(weight) / ((int(height)/100) ** 2))
        if bmi < 18.5:
            result = "저체중"
        elif 18.5 <= bmi <= 24.9:
            result = "정상체중"
        elif 25 <= bmi <= 29.9:
            result = "과체중"
        else:
            result = "비만"
        info_name.write("{},{}cm,{}kg,{},{}\n".format(name,height,weight,bmi,result))
        
with open("Info_name.txt","r") as info_name:
    for line in info_name:
        (name,height,weight,bmi,result) = line.strip().split(",")
        print('\n'.join([
            "이름 : {}",
            "키 : {}",
            "몸무게 : {}",
            "BMI : {}",
            "결과 : {}  \n"
        ]).format(name,height,weight,bmi,result))
```
