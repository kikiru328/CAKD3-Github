###### Module / Packages / Library
-  Module : 특정 기능들(함수,변수,클래스 등)이 구현되어 있는 파이썬 파일(.py)
-  Packages : Module들을 하나의 상위 폴더에 넣어놓은 것 
              Package를 표현해 주기 위해서 __init__.py 가 존재한다
-  Library : 여러 Module 과 Package 를 묶어 표현 함
             Python 설치시 기본적으로 설치되는 Library를 표준 Library라고 한다.
             외부 (3th party)에서 개발한 Module과 Package 를 묶어서 외부 Library 라고 함

```python
import random
answer = list('가나다라마바사아자차카타파하')
with open("Info_name.txt",'w') as info_name:
    for i in range(1000):
        name = random.choice(answer) + random.choice(answer)
        height = random.randrange(140,201)
        weight = random.randrange(40,151)
        bmi = round(int(weight) / ((int(height)/100) ** 2))
        if bmi < 18.5:
            result = "저체중"
        elif 18.5 <= bmi <= 24.9:
            result = "정상체중"
        elif 25 <= bmi <= 29.9:
            result = "과체중"
        else:
            result = "비만"
        info_name.write("{},{}cm,{}kg,{},{}\n".format(name,height,weight,bmi,result))
        
with open("Info_name.txt","r") as info_name:
    for line in info_name:
        (name,height,weight,bmi,result) = line.strip().split(",")
        print('\n'.join([
            "이름 : {}",
            "키 : {}",
            "몸무게 : {}",
            "BMI : {}",
            "결과 : {} \n"
        ]).format(name,height,weight,bmi,result))
```

```python
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
tt = sns.load_dataset('titanic')
tt.head()
```

```python
tt.info()
```

```python
tt.age.unique() #이러한 데이터는 그냥 쓸 수 없음.
                #범주화 필요 - categorization

#범주화

def age_category(age):
    category = ''
    if age <= 2: category = "Baby"
    elif 2 < age <= 12: category = "Child"
    elif 12 < age <= 18: category = "Teenager"
    elif 18 < age <= 35: category = "Young Adult"
    elif 35 < age <= 60: category = 'Adult'
    else: category = "Elderly"
    return category
        
tt['age_cat'] = tt['age'].apply(lambda x : age_category(x))
tt.head()
```

###### name == "main" 의미
- 메인 함수의 선언, 시작을 의미합니다. 해당 코드 밑에 main등의 함수 호출 코드를 작성해서 함수의 기능을 수행
- 직접 mod.py파일을 실행하는 경우에는 name_ 변수에 main이라는 값이 할당되고 이 조건절이 참이 되어 if문 다음 문장들이 수행
- 모듈을 불러와서 사용하는 경우 navme 변수에는 모듈 이름이 저장되고 이 조건절이 거짓이 되어 if문 다음 문장들이 수행되지 않는다.
- if name == "main"와 같이 조건문을 사용하여 터미널에서와 같이 직접 호출되어 사용될 때는 그 자체로 기능을 수행하고, 동시에 다른 모듈에서 필요한 함수 등을 제공한다.

```python
# long_task 함수는 수행하는데 5초의 시간이 걸리는 함수이다. 위 프로그램은 이 함수를 
#수행하는 프로그램이다
# 이 프로그램은 5초가 5번 반복되니 총 25초의 시간이 걸린다

import time
def long_task():
    for i in range(5):
        time.sleep(1) #1초대기
        print("working:%s\n"%i)
print('start')

for i in range(5): #long_task를 5회 수행
    long_task ()

print('End')
```

```python
import time
import threading

def long_task():
    for i in range(5):
        time.sleep(2) #1초대기
        print("working:%s\n"%i)
print('start')

threads = []
for i in range(5):
    t = threading.Thread(target=long_task)
    threads.append(t)
for t in threads:
    t.start()
for t in threads:
    t.join() # join 으로 thread가 종요할 때 까지 기다린다.
print('End')
```
