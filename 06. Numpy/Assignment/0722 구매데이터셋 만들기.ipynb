```python
# 구매 종류수
q2019 = abs(np.random.randn(1000)+5)
q2019 = q2019.round().astype(int)


# 매칭을 위한 DF
# 희소행렬..비슷하게?
price2019 = pd.DataFrame(index=range(1000), columns=range((100)))
qty2019 = pd.DataFrame(index=range(1000), columns=range((100)))

# 열인덱스가 좀.. 안좋지만 보기 좋게 담을 그릇
pid19 = []
qty19=[]

# 1000개의 인스턴스 데이터 생성
for i in range(1000):
    # 제품의 종류를 q2019[i]개만큼 선택함
    tempid = np.random.choice(100,q2019[i],replace=False)
    # 선택된 제품에 맞게 가격정보를 불러옴
    tempp = np.array(product.loc[tempid]['price19'].tolist() ).astype(int)
    # 제품의 종류별 (q2019[i]개 만큼의)  구매수량 생성
    tempq =  np.random.randint(1,101,q2019[i]).astype(int)
    
    # 구매 정보를 별도로 저장
    pid19.append(tempid)
    qty19.append( tempq )
    
    
    # 제품id와 매칭해서 저장
    price2019.loc[i,tempid] = tempp
    qty2019.loc[i,tempid] = tempq
    

# 꽁자제품의 가능성도 있으므로, 수량만 0을 넣어줌
# NaN이 있으면 int형이 불가능하므로 False를 주었음
# 
price2019= price2019.fillna(False)
qty2019= qty2019.fillna(0)

    
pid19 = pd.DataFrame(pid19).fillna(-1).astype(int).replace(-1,False)
qty19 = pd.DataFrame(qty19).fillna(-1).astype(int).replace(-1,False)
qty19

```

```python
# 구매 종류수
qall = abs(np.random.randn(인스턴스수)+5)
qall = qall.round().astype(int)


# 매칭을 위한 DF
# 희소행렬..비슷하게?
priceall = pd.DataFrame(index=range(인스턴스수), columns=range((100)))
qtyall = pd.DataFrame(index=range(인스턴스수), columns=range((100)))

# 열인덱스가 좀.. 안좋지만 보기 좋게 담을 그릇
pidmaster = []
qtymaster=[]

# 1000개의 인스턴스 데이터 생성
for i in range(인스턴스수):
    # 제품의 종류를 qall[i]개만큼 선택함
    tempid = np.random.choice(100,qall[i],replace=False)
    # 선택된 제품에 맞게 가격정보를 불러옴
    text = 'price' + str(alltime.iloc[i,1])
    tempp = np.array(product.loc[tempid][text].tolist() ).astype(int)
    # 제품의 종류별 (qall[i]개 만큼의)  구매수량 생성
    tempq =  np.random.randint(1,101,qall[i]).astype(int)
    
    # 구매 정보를 별도로 저장
    pidmaster.append(tempid)
    qtymaster.append( tempq )
    
    
    # 제품id와 매칭해서 저장
    priceall.loc[i,tempid] = tempp
    qtyall.loc[i,tempid] = tempq
    

    
priceall= priceall.fillna(False)
qtyall= qtyall.fillna(0)

    
pidmaster = pd.DataFrame(pidmaster).fillna(-1).astype(int).replace(-1,False)
qtymaster = pd.DataFrame(qtymaster).fillna(-1).astype(int).replace(-1,False)
qtymaster

```
