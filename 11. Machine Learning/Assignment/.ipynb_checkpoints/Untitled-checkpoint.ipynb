###### module import

```python
# basic module
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# warnings 
import warnings
warnings.filterwarnings('ignore')

# preprocessing


# seperate
from sklearn.model_selection import train_test_split

# machine learning
## Linear_Regression
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso, ElasticNet
## Trees
from sklearn.ensemble import RandomForestRegressor

# evaluation
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
```

# Data load

```python
house = pd.read_csv('Data/io_train.csv')
df = house.copy()

def data_info(df,null_count_standard):
    display(df.head(2))
    print('[shape] : ',df.shape)
    print('[rows] : {} , [columns] : {}'.format(df.shape[0],df.shape[1]),'\n')
    print('[house_value_counts] : ',df.dtypes.value_counts(),'\n')
    print('[columns] : ' ,df.columns,'\n')
    nl_v = df.isnull().sum()
    print('[Null_values] : ','\n',nl_v[nl_v>null_count_standard].sort_values(ascending=False))
    print('[Null_columns] : ','\n', list(nl_v[nl_v>null_count_standard].keys()))
data_info(df,500)
```

# Preprocessing 

## Preprocessing

##### 500개 이상 삭제

```python
# drop_columns
null_col = ['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
df = df.drop(null_col,axis=1)
df = df.drop('Id',axis=1)
```

##### 삭제하지 않은 숫자형 컬럼은 평균값으로 대체

```python
# replace to avg
df = df.fillna(df.mean())
```

```python
# check
data_info(df,0)
n_c = df.isnull().sum()[df.isnull().sum()>0].sort_values(ascending=False)
print()
print(df.dtypes[n_c.index])
```

##### 문자형 피처는 원-핫인코딩으로 변환

```python
df_oh = pd.get_dummies(df)
```

```python
def one_hot_check(df):
    print('[before] : ', df.shape)
    df_oe = pd.get_dummies(df)
    print('[after] : ', df_oe.shape)
    print('---' * 10)
    nl_cnt = df_oe.isnull().sum()[df_oe.isnull().sum()>0]
    print('[Type] : ', df_oe.dtypes[nl_cnt.index])
one_hot_check(df)
```

# EDA 

## Visulization

```python
# dependent variable
sns.distplot(house['SalePrice']) # 정규분포에서 벗어남\

# log Tf
log_sp = np.log1p(df['SalePrice'])
sns.displot(log_sp,kde=True)

# apply
df_oh['log_sp'] = np.log1p(df_oh['SalePrice'])
```

# Seperate

```python
X = df_oh.drop(['log_sp','SalePrice'],axis=1)
y = df_oh['log_sp']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=156)
```

# Machine Learning

## Model import

### Linear_Regression models

```python
lr = LinearRegression()
rdg = Ridge()
las = Lasso()
ela = ElasticNet()
```

### Regrssion trees models

# Fitting

```python
models = [lr,rdg,las,ela]
for model in models:
    model.fit(X_train,y_train)
```

# Predicting

```python
def varName(p):
    for k, v in globals().items():
        if id(p) == id(v):
            return k

for model in models:
    globals()[f'{varName(model)}_pred'] = model.predict(X_test)
```

# Evaluation

## 1st Evaluation

```python
def get_rmse(model):
    pred = model.predict(X_test)
    mse = mean_squared_error(y_test,pred)
    rmse = np.sqrt(mse)
    print('[model] : ',model.__class__.__name__)
    print('[rmsle] : ',np. round (rmse,5))
    print('---' * 20)
    return rmse
          
def get_rmses(models):
    rmses = [ ]
    for model in models:
        rmse = get_rmse(model)
        rmses.append(rmse)
    return 
```

```python
get_rmses(models)
```

## Coef_

```python
def get_coef(model,n=10):
    coef=pd.Series(model.coef_,index=X.columns)
    co_h = coef.sort_values(ascending=False).head(n)
    co_l = coef.sort_values(ascending=False).tail(n)
    return co_h, co_l
```

```python
def visual_coef(models):
    fig, axs = plt.subplots(figsize=(24, 10), nrows=1, ncols=4)
    fig.tight_layout()
    
    for i_num, model in enumerate(models):
        co_h, co_l = get_coef(model)
        coef_concat = pd.concat( [co_h, co_l] )
        axs[i_num].set_title(model.__class__.__name__ + 'Coeffiecents', size=25)
        axs[i_num].tick_params(axis="y", direction='in', pad= -120)
        for label in (axs[i_num].get_xticklabels() + axs[i_num].get_yticklabels()):
            label.set_fontsize(22)
        sns.barplot(x=coef_concat.values, y=coef_concat.index, ax=axs[i_num])
```

```python
models=[lr,rdg,las,ela]
visual_coef(models)
```

## 2nd Evaluation

##### 교차검증 수행 후 모델별 비교 - cross_val_score 이용

```python
def non_split_cv(models,cv_n):
    for model in models:
        rmse_ = np.sqrt(-cross_val_score(model,X,y,scoring='neg_mean_squared_error',cv=cv_n))
        avg_rmse = np.mean(rmse_)
        print(f'[model] : {model}')
        print(f'{cv_n} CV rmse : ', np.round(rmse_,3))
        print(f'{cv_n} CV avg_rmse : ',np.round(avg_rmse,3))
        print('---' * 20)
non_split_cv(models,5)        
```
