```python
# 0 부터 1000사이의 균일 분포를 이루는 난ㄴ수
np.random.seed(0)
data = np.random.rand(10000)*1000
print(data[:10])
df = pd.DataFrame(data, columns=['Num'])
display(df.describe())
df.plot(kind='hist')
```

```python
np.random.seed(3)
N_data = np.random.randn(10000)
ndf = pd.DataFrame(N_data, columns = ['Num'])
ndf.describe()
sns.histplot(x=ndf['Num'])
ndf.plot(kind='kde')
ndf.plot(kind='hist',bins=50)
```

```python
from scipy import stats
np.random.seed(2)
sample1 = stats.norm.rvs(loc=10,scale =2,size=10000)
df1 = pd.DataFrame(sample1, columns=['Num'])
print(df1.head())
print(df1.describe)
df1.plot(kind='hist',bins=50)
```

### Matplotlib

```python
x = np.arange(1,21,2)
y = np.array([3,5,6,13,30,21,22,29,33,32])

plt.figure(figsize=(14,4))

plt.plot(x,y,color='blue',marker='o',markersize=5)
plt.title('Lineplot Matplotlib',size=20)
plt.xlabel('x',size=15)
plt.ylabel('y',size=15)
plt.legend(labels=['lineplot'])
```

```python
np.random.seed(0)
x = np.arange(20)
y1 = np.arange(20) * 2 + np.random.randn(20) * 10
y2 = np.arange(20) * 2 + np.random.randn(20)

plt.figure(figsize=(10,6))
plt.plot(x,y1)
plt.plot(x,y2,color='magenta',marker='o',markersize=5)
plt.xlabel('x',size=20)
plt.ylabel('y',size=20)
plt.title('Line Graph',size=20)
plt.legend(['A(y1)','B(y2)'])
```

```python
#막대그래프
y = np.random.randint(2,10,size=10)
x = range(len(y))
print(x,y)
plt.figure(figsize=(8,4))
plt.bar(x,y,width=0.8,color='skyblue')
plt.ylabel('y')
plt.xlabel('x')
```
