```python
# 0 부터 1000사이의 균일 분포를 이루는 난ㄴ수
np.random.seed(0)
data = np.random.rand(10000)*1000
print(data[:10])
df = pd.DataFrame(data, columns=['Num'])
display(df.describe())
df.plot(kind='hist')
```

```python
np.random.seed(3)
N_data = np.random.randn(10000)
ndf = pd.DataFrame(N_data, columns = ['Num'])
ndf.describe()
sns.histplot(x=ndf['Num'])
ndf.plot(kind='kde')
ndf.plot(kind='hist',bins=50)
```

```python
from scipy import stats
np.random.seed(2)
sample1 = stats.norm.rvs(loc=10,scale =2,size=10000)
df1 = pd.DataFrame(sample1, columns=['Num'])
print(df1.head())
print(df1.describe)
df1.plot(kind='hist',bins=50)
```
